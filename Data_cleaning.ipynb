```python
from scipy.stats import chi2_contingency

# Define a function to perform and display Chi-Squared Test results
def perform_chi_squared_test(feature1, feature2, df):
    contingency_table = pd.crosstab(df[feature1], df[feature2])
    chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
    print(f"\nChi-squared Test: {feature1} vs. {feature2}")
    print(f"Chi-squared Statistic: {chi2_stat:.4f}, p-value: {p_value:.6f}, Degrees of Freedom: {dof}")
    print("Expected Frequencies:\n", expected)

# Perform Chi-Squared tests for IncidentGrade with EvidenceRole and EntityType
perform_chi_squared_test('IncidentGrade', 'EvidenceRole', df_Train)
perform_chi_squared_test('IncidentGrade', 'EntityType', df_Train)
```
